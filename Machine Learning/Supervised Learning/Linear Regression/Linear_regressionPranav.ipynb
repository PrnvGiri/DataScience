# What is VIF (Variance Inflation Factor) in Simple Terms?
VIF tells us **how much a particular independent variable (X) is correlated with other independent variables** in a regression model.

* If **VIF is high**, it means that the variable is highly correlated with others → **Multicollinearity problem** 🚨

* If VIF is low (close to 1), the variable is mostly independent → **Good for regression** ✅

##How to Interpret VIF?
* VIF = 1 → No correlation (Ideal)

* VIF < 5 → Low correlation (Okay)

* VIF > 10 → High correlation (Problematic, should be fixed!)
